```python
import pandas as pd
import os
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import shutil
```

```python
#Zmienne globalne
separator=';' # rzeczywisty separaor
index='id' # pierwsza kolumna która jest rzeczywistym indexem ( moze byc 0)
```

```python
# Dane wejsciowe
path_to_data='dane/' # folder odczytu wykresu

file1='fif_appar_0.csv' # plik z tabela do odczytu
file2='fif_appar_1.csv'
file3='fif_appar_2.csv'
file4='fif_appar_3.csv'
file5='fif_appar_4.csv'
file6='fif_appar_5.csv'
file7='fif_appar_6.csv'
file8='fif_appar_7.csv'


files=[file1,file2,file3,file4,file5,file6,file7,file8]
```

```python
# Dane dla tebeli
filtr_col='title' # szukana kolumna w wierszu dla którego chcemy zmineic dane
filtr_val='Punkt rosy/szronu' # szukane wyrazenie w wierszu dla ktorego chcemy zmienic dane
col_to_change='save_graph_data' # kolumna dla której chcemy zmienic dane
new_val=0 # nowa wartosc
```

```python
# Dane wyjsciowe
save_file_path=path_to_data+'new_data/' # docelowy folder zapisu zmienionego pliku
```

```python
def change_val(path,filtr_col,filtr_val,col_to_change,new_val):
    df=pd.read_csv(path, sep=separator, index_col='id') # pobranie tabeli raw
    print('Przed zmiana')
    print(df[[filtr_col,col_to_change]].head()) # wydruk tabeli przed zmiana
    filtr=df[filtr_col].str.contains(filtr_val,case=True) # ustalenie maski filtracyjnej
    df.loc[filtr,col_to_change]=new_val #zmiana wartosci w kolumnie po masce
    print('Po zmianie')
    print(df[[filtr_col,col_to_change]].head()) # wydruk tabeli po zmianie
    return df
```

```python
def zapis(save_file_path,file,df_do_zapisu):
    try:
        os.makedirs(save_file_path)
    except FileExistsError:
        print('Usuniecie istniejacego folderu: ', save_file_path)
        shutil.rmtree(save_file_path)  # usuniecie folderu kotory juz istnieje
        os.makedirs(save_file_path)
        pass
    print("Zapis do: ", save_file_path)
    try:
        df_do_zapisu.to_csv(save_file_path+file,sep=';') #zapis do pliku
    except Exception as e:
        print('Nie mozna zapisac pliku csv', e)
```

```python
for file in files:
    path=path_to_data+file
    dc=change_val(path,filtr_col,filtr_val,col_to_change,new_val) # zmiana wartosci
    zapis(save_file_path,file,dc) # zapis
    print('Po zapisie')
    dz=pd.read_csv(save_file_path+file, sep=separator,index_col='id') # odczyt zapisanego pliku
    print(dz[[filtr_col,col_to_change]].head())
    
```
